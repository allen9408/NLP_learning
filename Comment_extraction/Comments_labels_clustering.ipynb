```python
import pickle as pkl
comment_input_file = 'dzdp_hot_SH.txt'

comments = []
with open(comment_input_file, 'r') as f:
    for line in f:
        if line != '\n': comments.append(line)

len(comments)
```

```python
comment = comments[1]
comment
```

```python
import re
comment_to_list = re.split(' |,|，|。|;|\~', comment)
```

```python
import jieba.posseg as pseg
comment_cut_list = [pseg.cut(c) for c in comment_to_list]
for w, v in comment_cut_list[7]:
    print(w, v)
```

```python
word_vect_file = '/Users/allen/Codes/NLP_learning/Word_emb/bdbk_dict.pkl'
with open(word_vect_file, 'rb') as f:
    wrod_vect_dict = pkl.load(f)
```

```python
sentence_segments = []
for comment in comments:
    comment_to_list = re.split(' |,|，|。|;|\~|……|…|！|（|）|、|～|「|」|[|]', comment)
    sentence_segments += comment_to_list
len(sentence_segments)
```

```python
import jieba
sentence_word_lists = [list(jieba.cut(sen)) for sen in sentence_segments]
sentence_word_lists[0]    
```

```python
# Calculate hit prob
num_words, num_hits = 0, 0
for words_list in sentence_word_lists:
    for w in words_list:
        num_words += 1
        if w in wrod_vect_dict:
            num_hits += 1
print('Hit prob: ', num_hits/num_words)
```

```python
import numpy as np
def calculate_sentence_vector(word_dict, word_list):
#     print(word_list)
    word_vect_list = [word_dict[w] for w in word_list if w in word_dict]
    word_vect_array = np.array(word_vect_list)
    sentence_array = np.sum(word_vect_array, axis=0) / len(word_vect_list)
    sentence_value = sentence_array.tolist()
#     print(sentence_value)
    if type(sentence_value) != list or len(sentence_value) != 300:
        return [0] * 300
    return sentence_value
len(calculate_sentence_vector(wrod_vect_dict, sentence_word_lists[7]))
```

```python
sentence_vector_list = [calculate_sentence_vector(wrod_vect_dict, sentence) for sentence in sentence_word_lists]
```

```python
sentence_vector_list[0]
```

```python
from sklearn.cluster import KMeans
clusterer = KMeans(n_clusters=30, random_state=10)
cluster_labels = clusterer.fit_predict(sentence_vector_list)
```

```python
with open('cluster_results.txt', 'w') as f:
    
    for i in range(30):
        f.write(str(i) + '---------------------------------------' + '\n')
        sentences_to_write = [s for s, l in zip(sentence_segments, cluster_labels.tolist()) if l == i]
        for s in sentences_to_write:
            f.write(s + '\n')
```
