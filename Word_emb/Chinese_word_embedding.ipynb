```python
import pickle as pkl
with open('bdbk_dict.pkl', 'rb') as f:
    bdbk_dict = pkl.load(f)
```

```python
with open('demo.tsv', 'w') as f:
    thres = 10000
    count = 0
    first_line = 'word\t' + 'e\t'*300 + '\n'
    f.write(first_line)
    for k, v in bdbk_dict.items():
        line = ''
        line += k + '\t'
        line += '\t'.join(list(map(str,v)))
        line += '\n'
        f.write(line)
        count += 1
        if count == thres:
            break
        
```
